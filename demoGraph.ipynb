```python
#4-2 matplotlib라이브러리 자유자재로 사용하기
#기초 그래프 그리기
import seaborn as sns
tips = sns.load_dataset('tips')
# print(tips.head(5))
# print(type(tips))
# print(tips.shape)
# print(tips.info())
print(tips.tail(10))
```

```python
#변수 하나만 사용해서 그린 그래프는 일변량그래프라고 한다. 
import matplotlib.pyplot as plt 
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1) #그림판에 차트 하나
```

```python
#x축의 간격은 bins인자값으로 조정할 수 있다. 
#bins인자값을 10으로 지정하면 x축의 간격을 1으로 조정할 수 있다. 
axes1.hist(tips['total_bill'], bins=10) #hist 막대그래프
axes1.set_title('Hisogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')

fig
```

```python
#산점도 그래프는 변수 2개를 사용해서 만드는 그래프이며
#변수 2개를 사용하기 때문에 '이변량 그래프'라고 부른다. 
scatter_plot = plt.figure() 
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip']) #X축, Y축 - 2변량
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')            
```

```python
#박스그래프는 이산형변수와 연속형 변수를 함께 사용하는 그래프이다.
#이산형 변수란 Female, Male과 같이 명확하게 구분되는 값으 의미한다.
#연속형 변수란 Tip과 같이 명확하게 셀 수 없는 범위의 값을 의미한다. 

boxplot = plt.figure() 
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],
                tips[tips['sex'] == 'Male']['tip']],
                labels=['Female','Male'])
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

```python
#다변량 그래프 그리기
#3개 이상의 변수를 사용한 그래프는 '다변량 그래프'라고 한다. 

#Female, Male과 같은 문자는 그래프의 색상을 지정하는 값으로
#사용할 수가 없다. 그래서 0,1과 같은 숫자로 바꾸어야 한다. 
#성별에 따라 0과 1을 반환하도록 함수 정의 
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1 

#recode_sex함수를 브로드캐스팅하기 위해 apply()메서드 사용
#다중 행에 함수를 적용할 때 apply()메서드 사용 
tips['sex_color'] = tips['sex'].apply(recode_sex)

#테이블당 인원도 산점도에 추가 
#테이블당 인원 수는 점의 크기로 표현하면 적당하다.
#s는 점의 크기(size)를, c는 점의 색상(color)을 의미한다.  
scatter_plot = plt.figure() 
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size'] * 10,
    c=tips['sex_color'],
    alpha=0.5
)
axes1.set_title("Total Bill vs Tip Colored by Sex and Sized by Size")
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
#seaborn라이브러르를 사용하면 matplotlib보다 좀 더
#화려한 그래프를 그릴 수 있다.
import pandas as pd 
import matplotlib.pyplot as plt 
from matplotlib import font_manager, rc 
import seaborn as sns 

#한글 폰트롤 바꾸는 작업 
font_location='c:/Windows/Fonts/malgunsl.ttf'
font_name=font_manager.FontProperties(
    fname=font_location).get_name()
rc('font', family=font_name)


tips = sns.load_dataset('tips')

ax = sns.displot(tips['total_bill'])
```

```python
#displot메서드를 사용하면 히스토그램과 밀집도 그래프를 같이 그린다.
#만약 밀집도 그래프를 ㅔ외하고 싶다면 kde인자를 False로 지정한다. 
#밀집도 그래프는 주어진 데이터를 정규화시켜 넓이가 1이 되도록 그린 그래프이다.
ax = plt.subplots()
ax = sns.displot(tips['total_bill'], kde=True)
ax.set_titles('Total Bill Histogram')
ax.set_xlabels('Total Bill')
ax.set_ylabels('Frequency')
```

```python
ax = plt.subplots()
ax = sns.displot(tips['total_bill'], kde=False)
ax.set_titles('Total Bill Histogram')
ax.set_xlabels('Total Bill')
ax.set_ylabels('Frequency')
```

```python
#밀집도 그래프만 나타내려면 hist인자를 False로 지정한다.
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

```python
#실무 환경에서는 데이터 시각화를 더 효율적으로 하기 위해 
#여러 그래프를 한번에 출력하는 경우가 많다. 
#다음은 히스토그램과 밀집도를 그리는 distplot메서드에 rug인자를 
#추가하여 양탄자(rug)그래프까지 그린 것이다. 
#양탄자 그래프는 그래프의 축에 동일한 길이의 직선을 붙여
#데이터 밀집 정도를 표현한 그래프이다. 
ax = plt.subplots()
ax = sns.displot(tips['total_bill'], kde=True, rug=True)

ax.set_xlabels('Total Bill')
```

```python
#count그래프는 이산값(명확하게 구분되는 값)을 나타낸 그래프이다. 
#다음은 countplot메서드에 tips데이터프레임의 
#day열 데이터를 넣어 count그래프를 그린 것이다. 
ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

```python
#다양한 종류의 이변량 그래프 그리기
#산점도 그래프를 그리려면 seaborn라이브러리의 
#regplot메서드를 사용해야 한다. 
#regplot메서드를 사용하면 산점도 그래프와 회귀선을 함께 그릴 수 있다. 
#만약 회귀선을 제거하려면 fit_reg인자를 False로 지정한다. 
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
#회귀선을 제거하려면 fig_reg인자를 False값으로 지정한다.
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
#산점도 그래프와 히스토그램을 한번에 그려주는 jointplot메서드를
#사용해 본다. x,y인자에 원하는 열이름을 지정하고 
#data인자값으로 데이터프레임을 지정하면 된다. 

joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
#산점도 그래프는 점이 겹쳐 보일 경우 점을 구분하기 어렵다는 단점이 있다.
#육각 그래프I(hexbin)을 사용하면 된다. 
#육각 그래프는 2차원 표면에 육각형으로 데이터를 쌓아서 표현한다.
#그래서 특정 데이터의 개수가 많아지면 점점 진한 색으로 표현된다.
joint = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
#이차원 밀집도 그리기 
#이차원 밀집도는 kdeplot메서드로 만들 수 있다. 
#shade인자값을 True로 지정하면 그래프에 음영 효과를 줄 수 있다.
ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'],
    data2=tips['tip'],
    shade=True)

ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('tip')
```

```python
#바그래프 그리기
#바 그래프는 지정한 변수의 평균을 계산하여 그릴 수 있다. 
#다음은 시간에 따라 지불한 비용의 평균을 바 그래프로 나타낸 것이다.
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
#박스 그래프 그리기
#박스 그래프는 최소값, 1분위수, 중간값, 3분위수,
#최대값, 이상치 등 다양한 통계량을 한번에 표현하는 그래프이다.
ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel("Time of day")
ax.set_ylabel('Total Bill')
```

```python
#박스 그래프는 다양한 통계 수치를 확인하기 위해 자주 사용하는 그래프이다.
#하지만 데이터 분산이 모호하게 표현된다. 
#이런 경우 박스 그래프에 커널 밀도를 추정한 바이올린 그래프를 
#사용하면 된다.
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel("Time of day")
ax.set_ylabel('Total Bill')
```

```python
#관게 그래프 그리기
#관계 그래프는 지금까지 설명한 그래프를 종합한 그래프이다.
fig = sns.pairplot(tips)
```

```python
#이때 관계 그래프는 중복된 정보가 표현된다는 단점이 있따. 
#즉, 관계 그래프의 절반(대각선을 기준으로)은 나머지 절반과 같다. 
#이런 경우 중복된 그래프가 그려지는 위치를 직접 지정하여 원하는 
#그래프로 교체하면 된다. 
#map_upper메서드를 대각선을 기준으로 위쪽에 그릴 그래프를 지정한다.
#반대로 map_lower메서드는 대각선을 기준으로 아래쪽에 그릴 그래프를 지정한다.
#map_diag는 대각선을 중심으로 그래프를 그린다. 
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug=True)
plt.show()
```

```python
#4-5 seaborn라이브러리로 그래프 스타일 설정하기
#seaborn라이브러리로 만든 그래프의 스타일은 set_style메서드를 사용하여
#바꿀 수 있다. 
#seaborn라이브러리 스타일은 darkgrid, whitegrid, dark, white, ticks
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', 
    hue='sex', data=tips, split=True)
```

```python
#whitegrid로 스타일을 지정 
sns.set_style('whitegrid')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill',
    hue='sex', data=tips, split=True)
```

```python
#for문을 사용해서 모든 스타일을 하나씩 적용한 그래프
fig = plt.figure()

seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1 
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style)

fig.tight_layout()
```
